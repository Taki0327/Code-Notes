```python
import matplotlib.pyplot as plt
import networkx as nx
import pandas as pd
```

```python
data = pd.read_csv('./dataset/西游记三元组.csv',encoding="gbk")
SVOs=data.values.tolist()
```

```python
# 设置能够正常显示中文标签
plt.rcParams["font.sans-serif"]=["SimHei"]
# 设置绘图参数
fig = plt.figure(figsize = (12,8), dpi = 100)
# 创建一个图的对象
g_nx = nx.DiGraph()
```

```python
# 提取数据中的三元组关系
labels = {}
# 主谓宾
for subj, pred, obj in SVOs:
    g_nx.add_edge(subj, obj)
    labels[(subj, obj)] = pred
```

```python
# 生成关系图
pos = nx.spring_layout(g_nx, k = 5)
# 设置图中节点的圆形尺寸大小
nx.draw_networkx_nodes(g_nx, pos, node_color='pink', node_size = 
3000)
nx.draw_networkx_edges(g_nx, pos, width = 1)
# 画出节点的名字值
nx.draw_networkx_labels(g_nx, pos, font_size = 10, font_family = 
"sans-serif")
# 画出边的关系以及边的名字值
nx.draw_networkx_edge_labels(g_nx, pos, labels, font_size = 10, 
font_family = "sans-serif")
# 不展示坐标轴
plt.axis("off")
# 画出图谱
plt.show()
```
