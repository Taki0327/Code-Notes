# matplotlib实验

```python
import numpy as np,math,matplotlib.pyplot as plt
```

```python
x = np.arange(1,10)

y0 = np.sqrt(x)
y1 = np.power(x,2)
y2 = np.power(x,3)
y3 = np.log2(x)

plt.subplot(2,2,1)
plt.plot(x,y0)
plt.title("sqrt(x)")
plt.subplot(2,2,2)
plt.plot(x,y1)
plt.title("power(x,2)")
plt.subplot(2,2,3)
plt.plot(x,y2)
plt.title("power(x,3)")
plt.subplot(2,2,4)
plt.plot(x,y3)
plt.title("log2(x)")

plt.tight_layout()
plt.savefig('demo.png',dpi=300)
plt.show()
```

# 先规划再画图

```python
import numpy as np,math,matplotlib.pyplot as plt
```

```python
plt.rcParams['font.sans-serif'] = 'SimHei'

fig=plt.figure()
sub0=fig.add_subplot(2,2,1)
sub1=fig.add_subplot(2,2,2)
sub2=fig.add_subplot(2,2,3)
sub3=fig.add_subplot(2,2,4)

x = np.arange(1,10)
y0 = np.sqrt(x)
y1 = np.power(x,2)
y2 = np.power(x,3)
y3 = np.log2(x)

sub0.plot(x,y0)
sub0.set_title("sqrt(x)")
sub1.plot(x,y1)
sub1.set_title("power(x,2)")
sub2.plot(x,y2)
sub2.set_title("power(x,3)")
sub3.plot(x,y3)
sub3.set_title("log2(x)")
plt.show()
```

```python
import numpy as np,math,matplotlib.pyplot as plt
from skimage import data

img1=data.coffee()
img2=data.camera()
img3=data.coins()
img4=data.cell()

fig=plt.figure()
sub0=fig.add_subplot(2,2,1)
sub1=fig.add_subplot(2,2,2)
sub2=fig.add_subplot(2,2,3)
sub3=fig.add_subplot(2,2,4)

sub0.imshow(img1)
sub1.imshow(img2,cmap='gray')
sub2.imshow(img3)
sub3.imshow(img4)

plt.show()
```

```python
path_images=data.data_dir
print(path_images)
```

```python
import cv2
fig=plt.figure(figsize=(16,9))
index = 1
# sube = fig.add_ subplot(2,2,1)
for each in path_images:
    path=data.data_dir + "\{}".format(each)
    print(path)
    _img = cv2.imread(path)
    sub = fig.add_subplot(3, 6, index)
    if _img.ndim>2:
        sub.imshow(_img)
    else:
        sub.imshow(_img,cmap='gray')
    sub.set_title('{}'.format(each))
    index =index+1
plt.tight_layout()
plt.show()
```
