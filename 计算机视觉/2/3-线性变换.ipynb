```python
import numpy as np,matplotlib.pyplot as plt
from skimage import data

img=data.camera()
exp_img=np.copy(img)

alpha = 1
belta=[10,50,100]

fig=plt.figure(figsize=(16,9))
org=fig.add_subplot(2,2,1)
org.imshow(img,cmap='gray')
org.set_title("orginal image")

for index in range(3):
    for i in range(exp_img.shape[0]):
        for j in range(exp_img.shape[1]):
            _pixel = alpha * exp_img[i,j]+belta[index]
            if _pixel > 255:
                _pixel=255
            exp_img[i,j]=_pixel
    sub = fig.add_subplot(2,2,index+2)
    sub.imshow(exp_img,cmap='gray')
    sub.set_title('belta={}'.format(belta[index]))
plt.show()
```

```python
def linerTransfrom_gray(_img,_alpha,_belta):
    internal_img=np.copy(_img)
    for i in range(internal_img.shape[0]):
        for j in range(internal_img.shape[1]):
            _pixel = _alpha * internal_img[i,j]+_belta
            if _pixel > 255:
                _pixel = 255
            if _pixel < 0:
                _pixel = 0
            internal_img[i,j]=_pixel
    return internal_img
```

```python
img=data.coins()
fig=plt.figure(figsize=(21,9))
org=fig.add_subplot(3,3,1)
org.imshow(img,cmap="gray")
org.set_title("orginal image")

aplha=1
belta=[10,20,50,100,-10,-50,-100,-150]
for i in range(8):
    _img=linerTransfrom_gray(img,1,belta[i])
    sub=fig.add_subplot(3,3,i+2)
    sub.imshow(_img,cmap='gray')
    org.set_title("aplha=1,belta={}".format(belta[i]))
plt.tight_layout()
plt.show()
```

# OpenCV


```python
from skimage import data
import cv2 as cv
import matplotlib.pyplot as plt

img = data.camera()
img_1 = cv.convertScaleAbs(img,alpha=1,beta=10)
img_2 = cv.convertScaleAbs(img,alpha=1,beta=50)
img_3 = cv.convertScaleAbs(img,alpha=1,beta=100)

fig=plt.figure(figsize=(16,9))
org=fig.add_subplot(2,2,1)
org.imshow(img,cmap='gray')
org.set_title("orginal image")
_img_1=fig.add_subplot(2,2,2)
_img_1.imshow(img_1,cmap='gray')
_img_1.set_title("img_1 image")
_img_2=fig.add_subplot(2,2,3)
_img_2.imshow(img_2,cmap='gray')
_img_2.set_title("img_1 image")
_img_3=fig.add_subplot(2,2,4)
_img_3.imshow(img_3,cmap='gray')
_img_3.set_title("img_3 image")

plt.tight_layout()
plt.show()
```
