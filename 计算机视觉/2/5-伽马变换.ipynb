```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
from skimage import *
```

```python
def gamaTrans(_img,_gamma):
    internal_img=np.copy(_img)
    internal_img=internal_img / 255.0
    internal_img=np.power(internal_img,_gamma)
    
    internal_img=internal_img * 255
    internal_img=internal_img.astype(np.uint8)
    return internal_img

img=data.coffee()
gama=[0.1,0.4,0.6,0.8,1.5,1.8,1.9,2.0]
fig=plt.figure(figsize=(16,9))
sub = fig.add_subplot(3,3,1)
sub.imshow(img)
sub.set_title("orginal image")
for i in range(len(gama)):
    _img = gamaTrans(img,gama[i])
    sub=fig.add_subplot(3,3,i+2)
    sub.imshow(_img,cmap='gray')
    sub.set_title("gama={}".format(gama[i]))
plt.tight_layout()
plt.show()
```
