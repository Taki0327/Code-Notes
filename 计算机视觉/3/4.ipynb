```python
from skimage import data
import matplotlib.pyplot as plt
import numpy as np
img = data.astronaut()

test = np.zeros_like(img, dtype=np.uint8)
img_gray= np.zeros((img.shape[0], img.shape[1]),dtype=np.uint8)
for i in range(img. shape[0]):
    for j in range(img. shape[1]):
        img_gray[i,j] = np.dot(img[i,j,:], (0.3,0.59,0.11))
for i in range(img_gray.shape[0]):
    for j in range(img_gray.shape[1]):
        if img_gray[i,j] > 220:
            test[i,j]=(255,255,0)
        elif img_gray[i,j] > 200:
            test[i,j]=(255,0,255)
        elif img_gray[i,j] > 188:
            test[i,j]=(200,200,200)
        elif img_gray[i,j] > 150:
            test[i,j]=(100,255,100)
        elif img_gray[i,j] > 100:
            test[i,j]=(10,255,10)
        elif img_gray[i,j] > 88:
            test[i,j]=(100,100,100)
        elif img_gray[i,j] > 50:
            test[i,j]=(0,188,0)
        else:
            test[i,j]=(0,0,0)
fig = plt.figure(figsize=(21,9))
org = fig.add_subplot(1,3,1)
org.imshow(img)
org.set_title("orginal image")
_img = fig.add_subplot(1,3,2)
_img.imshow(img_gray,cmap='gray')
_img.set_title("_img image")
_test = fig.add_subplot(1,3,3)
_test.imshow(test,cmap='gray')
_test.set_title("test image")
plt.tight_layout()
plt.show()
```
