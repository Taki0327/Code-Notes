```python
from skimage import data
import matplotlib.pyplot as plt
import numpy as np

img = data.astronaut()
max_img = np.zeros((img.shape[0], img.shape[1]),dtype=np.uint8)
avg_img= np.zeros((img.shape[0], img.shape[1]),dtype=np.uint8)
weight_img= np.zeros((img.shape[0],img.shape[1]),dtype=np.uint8)
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
       max_img[i,j] = np.max(img[i,j,:])
       avg_img[i,j] = np.mean(img[i,j,:])
       weight_img[i,j] = np.dot(img[i,j,:],(0.3,0.59,0.11))#dot点集
fig = plt.figure(figsize=(21,9))
org = fig.add_subplot(2,2,1)
org.imshow(img)
org.set_title("orginal image")
_max = fig.add_subplot(2,2,2)
_max.imshow(max_img,cmap='gray')
_max.set_title("max image")
_mean = fig.add_subplot(2,2,3)
_mean.imshow(avg_img,cmap='gray')
_mean.set_title("avg_img image")
_weight = fig.add_subplot(2,2,4)
_weight.imshow(weight_img,cmap='gray')
_weight.set_title("weight_img image")

plt.tight_layout()
plt.show()
```
