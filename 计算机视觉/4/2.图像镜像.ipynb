# 图像镜像 

```python
from skimage import data
import numpy as np,matplotlib.pyplot as plt

def mirror(img,category):
    height=img.shape[0]# y
    width=img.shape[1]# x
    if category==1:
        matrix=np.array([[-1,0,width],[0,1,0],[0,0,1]])#水平
    else:
        matrix=np.array([[1,0,0],[0,-1,height],[0,0,1]])#垂直
    test=np.zeros_like(img,dtype=np.uint8)
    for i in range(height):
        for j in range(width):
            org_coord = np.array([i,j,1])
            new_matrix1=matrix@org_coord.T #@矩阵乘法 .T转置 
            if new_matrix1[0]>=0 and new_matrix1[0]<height and new_matrix1[1]>=0 and new_matrix1[1]<width:
                test[new_matrix1[0],new_matrix1[1]]=img[i,j]
    return test
```

```python
img = data.camera()
fig = plt.figure(figsize=(21,9))
org = fig.add_subplot(1,3,1)
org.imshow(img,cmap='gray')
org.set_title("orginal image")

_test = fig.add_subplot(1,3,2)
_test.imshow(mirror(img,1),cmap='gray')#水平
_test.set_title("test image")

_test2 = fig.add_subplot(1,3,3)
_test2.imshow(mirror(img,2),cmap='gray')#垂直
_test2.set_title("test2 image")

plt.tight_layout()
plt.show()
```
