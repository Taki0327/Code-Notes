```python
from myplot import *
import numpy as np
import matplotlib.pyplot as plt
from skimage import data,io,feature
```

```python
imglist=[]
def canny(pic,param):
    edges1 = feature.canny(pic,sigma=param)
    imglist.append((edges1,str(param)))
```

```python
img = data.camera()

imglist.append((img,'原始图像'))
canny(img,3)
canny(img,5)
canny(img,7)
canny(img,9)

myPlot(imglist)
```
